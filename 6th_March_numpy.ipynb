## 2. Array Manipulation Functions

```python
a1=np.array([1,2,3])
reshaped=np.reshape(a1, (3,1))
print('Reshapd array',reshaped)
```

```python
f1=np.array([[1,2],[3,4]])
flattened=np.ravel(f1)
print("Flattened arrray", flattened)
```

```python
e1=np.array([[1,2],[3,4]])
transposed=np.transpose(e1)
print("Transposed arrray:\n", transposed)
```

```python
a2=np.array([1,2])
b2=np.array([3,4])
stacked=np.vstack([a2,b2])
print("Stacked arrays:\n",stacked)
```

```python
a2=np.array([1,2])
b2=np.array([3,4])
stacked=np.hstack([a2,b2])
print("Stacked arrays:\n",stacked)
```
